## Q1 Adjacency Matrix

```python
import numpy as np
import pandas as pd

adj_mat = np.zeros(shape = (1005, 1005), dtype = int)
with open("email-Eu-core.txt", "r") as file:
    f = file.readlines()
    for i in f:
        lis = i.split(" ")
        adj_mat[int(lis[0])][int(lis[1])] = 1

for i in adj_mat:
    print(i)
```

## Edge List

```python
edge_list = {}
with open("email-Eu-core.txt", "r") as file:
    f = file.readlines()
    for i in f:
        lis = i.split(" ")
        if int(lis[0]) in edge_list:
            edge_list[int(lis[0])].append(int(lis[1]))
        else:
            ls = []
            ls.append(int(lis[1]))
            edge_list[int(lis[0])] = ls
        
        
sorted_edge_list = dict(sorted(edge_list.items(), key=lambda x: x[0]))

for key, value in sorted_edge_list.items():
    print(key, sorted(value))
```

## Part 1 Number of Nodes

```python
num_of_node = 0
node_set = set()
with open("email-Eu-core.txt", "r") as file:
    f = file.readlines()
    for i in f:
        lis = i.split(" ")
        node_set.add(int(lis[1]))
        node_set.add(int(lis[0]))
num_of_node = len(node_set)
print(num_of_node)
```

## Part 2 Number of Edges

```python
edge_count = 0
with open("email-Eu-core.txt", "r") as file:
    f = file.readlines()
    edge_count = len(f)
print(edge_count)
```

## Part 3 Average In-Degree

```python
def findInDegree(sorted_edge_list, n):
     
    in_deg = np.zeros(n)
 
    for i in range(0, len(sorted_edge_list)):

        if i not in sorted_edge_list:
            continue
        List = sorted_edge_list[i]
        for j in range(0, len(List)):
 
            in_deg[List[j]] += 1
 
    return in_deg

in_deg = findInDegree(sorted_edge_list, len(node_set))
print(in_deg)
avg_in_deg = np.sum(in_deg)/len(in_deg)
print(avg_in_deg)
```

## Part 4 Average Out-Degree

```python
out_deg = np.zeros(len(node_set))
for key, value in sorted_edge_list.items():
    out_deg[key] = len(value)
print(out_deg)
avg_out_deg = np.sum(out_deg)/len(out_deg)
print(avg_out_deg)
```

## Part 5 Node with Max In-Degree

```python
ma = np.max(in_deg)
print(ma)
print(np.where(in_deg == ma)[0])
```

## Part 6 Node with Max Out-Degree

```python
ma = np.max(out_deg)
print(ma)
print(np.where(out_deg == ma)[0])
```

## Part 7 Density of the network

```python
density = edge_count/(num_of_node-1)*num_of_node
print(density)
```

## Part 8 Degree Distribution of Network

```python
import matplotlib.pyplot as plt

in_degree = {}
with open("email-Eu-core.txt", "r") as file:
    f = file.readlines()
    for i in f:
        lis = i.split(" ")
        a = int(lis[1])
        if a not in in_degree:
            in_degree[a] = 1
        else:
            in_degree[a] += 1
            
in_deg_count = {}

for deg in in_degree.values():
    if deg not in in_deg_count:
        in_deg_count[deg] = 1
    else:
        in_deg_count[deg] += 1
    
x = in_deg_count.keys()
y = in_deg_count.values()
plt.bar(x, y)
plt.xlabel('In-degree')
plt.ylabel('Count')
plt.title('In-degree Distribution')
plt.show()
```

```python
out_degree = {}
with open("email-Eu-core.txt", "r") as file:
    f = file.readlines()
    for i in f:
        lis = i.split(" ")
        a = int(lis[0])
        if a not in out_degree:
            out_degree[a] = 1
        else:
            out_degree[a] += 1
            
out_deg_count = {}

for deg in out_degree.values():
    if deg not in out_deg_count:
        out_deg_count[deg] = 1
    else:
        out_deg_count[deg] += 1
    
x = out_deg_count.keys()
y = out_deg_count.values()
plt.bar(x, y)
plt.xlabel('Out-degree')
plt.ylabel('Count')
plt.title('Out-degree Distribution')
plt.show()
```

## Part 9 Local Clustering Coefficient

```python
import matplotlib.pyplot as plt

def read_graph_from_file(file_path):
    adjacency_list = {}
    with open(file_path, "r") as file:
        lines = file.readlines()
        for line in lines:
            u, v = map(int, line.split())
            adjacency_list.setdefault(u, set()).add(v)
            adjacency_list.setdefault(v, set()).add(u)
    return adjacency_list

def calculate_clustering_coefficient(adjacency_list, node):
    neighbors = adjacency_list[node]
    if len(neighbors) < 2:
        return 0
    edges_within_neighbors = 0
    for u in neighbors:
        for v in neighbors:
            if u != v and v in adjacency_list[u]:
                edges_within_neighbors += 1
    return 2 * edges_within_neighbors / (len(neighbors) * (len(neighbors) - 1))

def calculate_clustering_coefficient_distribution(adjacency_list):
    clustering_coefficients = [calculate_clustering_coefficient(adjacency_list, node) for node in adjacency_list]
    clustering_coefficient_counts = {}
    for coefficient in clustering_coefficients:
        clustering_coefficient_counts[coefficient] = clustering_coefficient_counts.get(coefficient, 0) + 1
    return clustering_coefficient_counts


adjacency_list = read_graph_from_file("email-Eu-core.txt")


clustering_coefficient_counts = calculate_clustering_coefficient_distribution(adjacency_list)


plt.bar(clustering_coefficient_counts.keys(), clustering_coefficient_counts.values())
plt.xlabel('Clustering Coefficient')
plt.ylabel('Count')
plt.title('Clustering Coefficient Distribution')
plt.show()
```

## Q2 Part 1 PageRank score

```python
import networkx as nx

edges = list()
with open("email-Eu-core.txt", "r") as file:
    f = file.readlines()
    for i in f:
        lis = i.split(" ")
        edges.append((int(lis[0]), int(lis[1])))

G = nx.DiGraph()
G.add_edges_from(edges)

score = nx.pagerank(G)

for key, value in score.items():
    print(key, " : ",  value)
```

## Q2 Part 2 Authority and Hub score

```python
a, b = nx.hits(G)
print("node    Score          Hub Score")
for key, value in a.items():
    print(key, value, b[key])
```

## Q2 Part 3 Comparing Result

```python
import numpy as np
pgr, auth, hub = np.zeros(1005), np.zeros(1005), np.zeros(1005)
for key, val in b.items():
    pgr[key] = score[key]
    auth[key] = a[key]
    hub[key] = val

corr_coef_auth = np.corrcoef(pgr, auth)[0][1]
corr_coef_hub = np.corrcoef(pgr, hub)[0][1]

print("Pearson correlation coefficient between PageRank Score and Authority score: ",corr_coef_auth)
print("Pearson correlation coefficient between PageRank Score and Hub score: " ,corr_coef_hub)
```

```python
import matplotlib.pyplot as plt

dict1 = score
dict2 = a
dict3 = b

labels = list(dict1.keys())

bar_width = 0.2

fig, ax = plt.subplots()
ax.bar(labels, list(dict1.values()), width=bar_width, label='PageRank Score')
ax.bar([i + bar_width for i in range(len(labels))], list(dict2.values()), width=bar_width, label='Hub Score')
ax.bar([i + 2*bar_width for i in range(len(labels))], list(dict3.values()), width=bar_width, label='Authority Score')

ax.set_xlabel('Nodes')
ax.set_ylabel('Values')

ax.legend()

plt.show()
```
